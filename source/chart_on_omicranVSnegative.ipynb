```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv('../Original dataset/Omicran_vs_Negative.csv', low_memory=False)
def filter(f):
    columns = ['binaryclassification']
    x = ds[(ds['binaryclassification']==f)]
    x = x.drop(columns, axis = 1)
    x = x.iloc[:]
    return x.mean()
```

```python
days = [1, 0]
color = ['r', 'g']
colors = {'Positive (25pg/mL)':'red', 'Negative':'green'}

labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]

plt.figure(figsize=(16,8))
plt.xlabel("Wavenumber ($\mathregular{cm^{-1}}$)")
plt.ylabel('Intensity (a.u.)')

for f in range(0, len(days)):
    x = filter(days[f])
    plt.plot(x.index, x.values, color=color[f], linestyle = '-', linewidth=2)
    if f == 0:
        plt.xticks(plt.xticks()[0][1::50])
        plt.legend(handles, labels)
```

```python
fig = plt.figure(figsize=(8,6))
# fig.suptitle('Patient 2', fontsize=18)
gs = fig.add_gridspec(2, hspace=0.05)
axs = gs.subplots(sharex=True, sharey=True)

fig.text(0.06, 0.5, 'Intensity (a.u.)', ha='center', va='center', rotation='vertical')
fig.text(0.5, 0.04, 'Wavenumber ($\mathregular{cm^{-1}}$)', ha='center', va='center')
axs[0].text(0.48, 0.7, '1078 $\mathregular{cm^{-1}}$', color='black', fontsize=10,ha='center', va='bottom', transform=axs[0].transAxes)
axs[0].text(0.83, 0.7, '1582 $\mathregular{cm^{-1}}$', color='black', fontsize=10,ha='center', va='bottom', transform=axs[0].transAxes)
for f in range(0, len(days)):
    x = filter(days[f])
    axs[f].plot(x.index, x.values, color=color[f], linestyle = '-', linewidth=2, label = labels[f])
    axs[f].legend(loc="upper left")
    
    axs[f].axvline(x=756, color='black', linestyle='--', linewidth=1) # 1078
    axs[f].axvline(x=1260, color='black', linestyle='--', linewidth=1) # 1582    
    

    if f == 0:
        plt.xticks(plt.xticks()[0][1::100])
```
