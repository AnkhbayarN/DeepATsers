```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv('../Original dataset/Omicran_sample_concentration.csv', low_memory=False)
def filter(f):
    columns = ['concentration']
    x = ds[(ds['concentration']==f)]
    x = x.drop(columns, axis = 1)
    x = x.iloc[:]
    return x.mean()
```

```python
days = ['1ngml', '1ugml', '10ngml', '25pgml', '50pgml', '100ngml', '100pgml']
# color = ['g', 'b', 'k', 'm', 'r']
color = ['r', 'r', 'r', 'r', 'r', 'r', 'r']
colors = {'1Ng/mL':'green', '1ug/mL':'blue', '10Ng/mL':'black', '25pg/mL':'magenta', '50pg/mL':'red', '100Ng/Ml':'red', '100pg/mL':'red'}

labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
```

```python
plt.figure(figsize=(16,8))
plt.xlabel("Wavenumber ($\mathregular{cm^{-1}}$)")
plt.ylabel('Intensity (a.u.)')

for f in range(0, len(days)):
    x = filter(days[f])
    plt.plot(x.index, x.values, color=color[f], linestyle = '-', linewidth=2)
    if f == 0:
        plt.xticks(plt.xticks()[0][1::50])
        plt.legend(handles, labels)
```
